```python
import numpy as np 
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
def get_df():
    mental_health_df = pd.read_csv("Mental Health Dataset.csv")
    filtered_mental_health_df = mental_health_df.dropna()
    print(filtered_mental_health_df['care_options'].value_counts())
    return filtered_mental_health_df

mental_health_df = get_df()
```

```python
get_df()
```

```python
def clean_data(mental_health_df):
     mental_health_df['Gender'] = np.where(mental_health_df['Gender'] == 'Male,', 1, 0)
     mental_health_df['self_employed'] = np.where(mental_health_df['self_employed'] == 'Yes,', 1, 0)
     mental_health_df['family_history'] = np.where(mental_health_df['family_history'] == 'Yes,', 1, 0)
     mental_health_df['treatment'] = np.where(mental_health_df['family_history'] == 'Yes,', 1, 0)
     mental_health_df['Coping_Struggles'] = np.where(mental_health_df['Coping_Struggles'] == 'Yes,', 1, 0)


     mental_health_df["Growing_Stress"]=np.where(mental_health_df["Growing_Stress"]=="Yes",1,
                                    np.where(mental_health_df["Growing_Stress"]=="No",0,
                                             2))
     mental_health_df["Changes_Habits"]=np.where(mental_health_df["Changes_Habits"]=="Yes",1,
                                    np.where(mental_health_df["Changes_Habits"]=="No",0,
                                             2))
     mental_health_df["Mental_Health_History"]=np.where(mental_health_df["Mental_Health_History"]=="Yes",1,
                                    np.where(mental_health_df["Mental_Health_History"]=="No",0,
                                             2))
     mental_health_df["Work_Interest"]=np.where(mental_health_df["Work_Interest"]=="Yes",1,
                                    np.where(mental_health_df["Work_Interest"]=="No",0,
                                             2))
     mental_health_df["Social_Weakness"]=np.where(mental_health_df["Social_Weakness"]=="Yes",1,
                                    np.where(mental_health_df["Social_Weakness"]=="No",0,
                                             2))
     mental_health_df["mental_health_interview"]=np.where(mental_health_df["mental_health_interview"]=="Yes",1,
                                    np.where(mental_health_df["mental_health_interview"]=="No",0,
                                             2))
     mental_health_df["care_options"]=np.where(mental_health_df["care_options"]=="Yes",1,
                                    np.where(mental_health_df["care_options"]=="No",0,
                                             2))


     occupation_map = {'Housewife':1,'Student':2, 'Corporate':3, 'Others':4, 'Business':5}
     
     mental_health_df['Occupation'] = mental_health_df['Occupation'].map(occupation_map)

     days_indoors_map = {'1-14 days': 1, '31-60 days':3, 'Go out Every day': 5, 'More than 2 months':4, '15-30 days':2}
     mental_health_df['Days_Indoors'] = mental_health_df['Days_Indoors'].map(days_indoors_map)

     mood_swings_map = {'Low': 1, 'Medium':2, 'High': 3}
     mental_health_df['Mood_Swings'] = mental_health_df['Mood_Swings'].map(mood_swings_map)


     class_labels = LabelEncoder()
     mental_health_df['Country'] = class_labels.fit_transform(mental_health_df['Country'].values)
     return mental_health_df
```

```python
clean_data(mental_health_df).to_csv("cleaned_mental_health_data.csv")
```
